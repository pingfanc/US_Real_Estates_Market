### Transaction - join tables
The first table stores the US economic index, including GDP, unemployment rate, inflation rate, and average income. There are lots of missing value in the table because the data of inflation rate and income doesn't start from 1947. However, if we inner join every table, we will loss lots of information. 

```python
# join the tables of gdp and unemployment rate
joined_us_eco = cleaned_ps_df_gdp.join(cleaned_ps_df_unemployment, cleaned_ps_df_gdp.date == cleaned_ps_df_unemployment.date, "left")
joined_us_eco = joined_us_eco.drop(cleaned_ps_df_unemployment.date)

# join the joined table with the inflation rate table
joined_us_eco = joined_us_eco.join(cleaned_ps_df_inflation, joined_us_eco.date == cleaned_ps_df_inflation.date, "outer")
joined_us_eco = joined_us_eco.drop(cleaned_ps_df_inflation.date)

# join the joined table with the average incame table
joined_us_eco = joined_us_eco.join(cleaned_ps_df_des_income, joined_us_eco.date == cleaned_ps_df_des_income.date, "left")
joined_us_eco = joined_us_eco.drop(cleaned_ps_df_des_income.date)
joined_us_eco.show()
```

```python
# import the dataframe into postG
joined_us_eco.write.jdbc(url=postgres_uri, table="schema_enpeixia23uclacuk.us_eco", mode="overwrite", properties=properties)
```

The second combined data is about house market in the State, including the permits of construction, the home sales in the US, All-transactions house price index, and the CPI for urban rent. We keep all of the missing value of sales and CPI because the data only start from these two year. Removing all of them will lost massive data from other tables.

```python
# join the tables of gdp and unemployment rate
joined_us_estate = cleaned_ps_df_cpi_rent.join(cleaned_ps_df_building_permits, cleaned_ps_df_cpi_rent.date == cleaned_ps_df_building_permits.date, "left")
joined_us_estate = joined_us_estate.drop(cleaned_ps_df_building_permits.date)

joined_us_estate = joined_us_estate.join(cleaned_ps_df_home_sales, joined_us_estate.date == cleaned_ps_df_home_sales.date, "outer")
joined_us_estate = joined_us_estate.drop(cleaned_ps_df_home_sales.date)

joined_us_estate = joined_us_estate.join(cleaned_ps_df_home_price_index, joined_us_estate.date == cleaned_ps_df_home_price_index.date, "outer")
joined_us_estate = joined_us_estate.drop(cleaned_ps_df_home_price_index.date)


joined_us_estate.show()
```

```python
# import the dataframe into postG
joined_us_estate.write.jdbc(url=postgres_uri, table="schema_enpeixia23uclacuk.us_estate", mode="overwrite", properties=properties)
```

The third joined table is about the interest rate, including the interest rate of multu-familiy estate, commercial estate, and owner occupied estate.

```python
# join the tables of gdp and unemployment rate
joined_us_interest = cleaned_ps_df_interest_family.join(cleaned_ps_df_interest_commercial, cleaned_ps_df_interest_family.date == cleaned_ps_df_interest_commercial.date, "left")
joined_us_interest = joined_us_interest.drop(cleaned_ps_df_interest_commercial.date)

joined_us_interest = joined_us_interest.join(cleaned_ps_df_interest_owner_occupied, joined_us_interest.date == cleaned_ps_df_interest_owner_occupied.date, "left")
joined_us_interest = joined_us_interest.drop(cleaned_ps_df_interest_owner_occupied.date)

joined_us_interest.show()
```
